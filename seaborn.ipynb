```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
x=[1,2,3,4,5,6,5,6]
y=[10,20,30,40,50,60,1000,2000]
z=[12,13,14,15,16,17,199,200]
gen=['f','m','f','f','f','m','m','f']
df=pd.DataFrame([x,y,z,gen],index=['x','y','z','gender']).T
df
```

```python
plt.figure(figsize=(5,5))
sns.lineplot(data=df,x='x',y='y',hug='gender')
plt.xlable('x value')
plt.ylable('y value')
plt.title('graph between x and  z  with res to zender')
plt.show()
```

```python
sns.lineplot(data=df,x='x',y='y',hue='gender')
```

```python
sns.boxenplot(data=df,x='x',y='y',hue='gender',linewidth=1.5,linecolor='b')
```
