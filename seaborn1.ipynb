Bar plot in seaborn liberary

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
df1['island'].unique()
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm,)
plt.show()
```

```python
df1
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex')
plt.title("Bill Length")
plt.show()
```

Order parameter

```python
order1= ['Dream','Lorgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order1,hue='sex')
plt.show()
```

hue_order

```python
order1= ['Dream','Lorgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order1,
            hue='sex',hue_order=['Female','Male'])
plt.show()
```

```python
order1 =['Dream','Lorgerson','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order1,hue='sex',color='b')
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',palette='Accent')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',palette='prism_r')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',
            palette='gist_heat_r',saturation=0.1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot  as plt
import pandas as pd
```

```python
df1
```

```python
sns.histplot(df1['flipper_length_mm'])
```

bins

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

Color

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],color='g')
plt.show()
```

histogram Kde

```python
sns.displot(df1['flipper_length_mm'],bins =[170,180,190,200,210,220,230,240],kde =True,color='red')
plt.show()
```
